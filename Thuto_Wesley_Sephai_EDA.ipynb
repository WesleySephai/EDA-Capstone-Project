```python
#11. Use a countplot graph to display the number of male and female penguins
#Give the plot a title "Count of male and female penguins"
#Give a xlabel and ylabel 
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='sex', data=df)

plt.title('Count of Male and Female Penguins')
plt.xlabel('Sex')
plt.ylabel('Count')

plt.show()
```

```python
#13.Create a visualization to display the flipper_length_mm and body_mass_g for the penguin dataset.
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='flipper_length_mm', y='body_mass_g', data=df)
plt.title('Scatter Plot of Flipper Length vs Body Mass')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Body Mass (g)')
plt.show()


plt.figure(figsize=(8, 6))
sns.regplot(x='flipper_length_mm', y='body_mass_g', data=df)
plt.title('Regression Plot of Flipper Length vs Body Mass')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Body Mass (g)')
plt.show()
```
